# 1. Data Preprocessing#

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# The dataset
df_training = pd.read_csv("Training (1).csv")

# 1. Handle Missing Values
# Drop rows with missing values
df_training.dropna(inplace=True)

# 2. Encode Categorical Variables
# 'prognosis' is the only non-numeric column that needs encoding
le = LabelEncoder()
df_training['prognosis'] = le.fit_transform(df_training['prognosis'])

# 3. Normalize the Data
# All features except 'prognosis' should be normalized
scaler = MinMaxScaler()
cols_to_normalize = df_training.columns.difference(['prognosis'])
df_training[cols_to_normalize] = scaler.fit_transform(df_training[cols_to_normalize])

# Save the cleaned dataset
df_training.to_csv("Cleaned_Training.csv", index=False)
df_training.head(10)
```

# c. Data Categorization Based on Symptom Occurrence

```python
import pandas as pd
# Symptom columns: all except 'prognosis'
symptom_columns = df_training.columns.drop('prognosis')

# Calculate the mean for each symptom column
mean_values = df_training[symptom_columns].mean()

# Define the quartile thresholds
quantiles = mean_values.quantile([0.25, 0.5, 0.75]).values

# Function to assign categories and weights based on the mean occurrence
def categorize_occurrence(mean):
    if mean > quantiles[2]:
        return 'Very Common', 1
    elif mean > quantiles[1]:
        return 'Common', 2
    elif mean > quantiles[0]:
        return 'Uncommon', 3
    else:
        return 'Rare', 4

# Apply the categorization to every symptom
categories_weights = mean_values.apply(categorize_occurrence)

# Create a new DataFrame for the categorized data
categorized_data = pd.DataFrame(categories_weights.tolist(), index=mean_values.index, columns=['Category', 'Weight'])

categorized_data['Mean Occurrence'] = mean_values

categorized_data.sort_values(by=['Weight', 'Mean Occurrence'], ascending=[True, False], inplace=True)

print(categorized_data)
```

```python
def weight_color(val):
    color = 'black'
    if val == 1:
        color = 'green'
    elif val == 2:
        color = 'blue'
    elif val == 3:
        color = 'orange'
    else:
        color = 'red'
    return f'color: {color}'

styled_table = categorized_data.style.applymap(weight_color, subset=['Weight'])\
                                     .set_properties(**{'text-align': 'left'})\
                                     .set_table_styles([dict(selector='th', props=[('text-align', 'left')])])\
                                     .set_caption("Symptoms Categorized by Mean Occurrence with Weights")\
                                     .background_gradient(cmap='viridis', subset=['Mean Occurrence'])\
                                     .format({'Mean Occurrence': "{:.2f}"})


styled_table
```

# 3. Feature Weighting and Model Training

#a. Model Evaluation

The purpose of this weighting is to adjust the importance of each symptom in the dataset based on its frequency of occurrence. Common symptoms are given less weight because they might be less discriminative for diagnosing different diseases (since they occur in many different diseases), while rarer symptoms are given more weight since they might be more indicative of specific diseases.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, learning_curve
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# The features and target variable
X = df_training[symptom_columns]
y = df_training['prognosis']


# Apply MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Feature Selection based on feature importance
selector = SelectFromModel(RandomForestClassifier(n_estimators=50))
X_reduced = selector.fit_transform(X_scaled, y)

# Stratified Split
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42, stratify=y)

# Simplified RandomForest model
model = RandomForestClassifier(n_estimators=30, max_depth=8, random_state=42)
model.fit(X_train, y_train)

# Evaluate Feature Importance
importances = model.feature_importances_

# Visualize feature importances

# Cross-Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_train, y_train, cv=cv)

# Learning Curve Analysis
train_sizes, train_scores, test_scores = learning_curve(model, X_train, y_train, cv=cv, n_jobs=-1,train_sizes=np.linspace(.1, 1.0, 5))

# Plot learning curve
plt.figure()
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', color="r", label="Training score")
plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', color="g", label="Cross-validation score")
plt.title("Learning Curve")
plt.xlabel("Training Examples")
plt.ylabel("Score")
plt.legend(loc="best")
plt.show()

# Evaluate the model
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)

# Results
print(f"Cross-Validated Training Accuracy: {np.mean(cv_scores)}")
print(f"Test Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import difflib
from sklearn.feature_selection import SelectFromModel


def predict_disease(user_symptoms, symptoms_dict, model, selector):
    input_vector = np.zeros(len(symptoms_dict))
    for symptom in user_symptoms:
        # Check if the symptom is in our symptoms dictionary
        if symptom in symptoms_dict:
            # Get the index from the dictionary
            index = symptoms_dict[symptom]
            # Set the position in the input_vector to 1 when the symptom is present
            input_vector[index] = 1

    # Transform the input vector (patient's symptoms) using the feature selector
    input_vector_transformed = selector.transform([input_vector])

    # Predict the disease index based on the input vector
    predicted_index = model.predict(input_vector_transformed)[0]

    # Return the predicted disease index
    return predicted_index
```
